#### Задача 1
Напишите функцию, которая будет переводить `int` в `str` без использования функции `str()`.

- без форматирования строки;

````python
const = 48  # const = ord('0')

def int1_to_str(number: int) -> str:
    return chr(number + const)

def int_to_str(number: int) -> str:
    a = abs(number)

    if number != 0:
        result = ''
        
        while a != 0:  # a > 0
            num = a % 10
            result += int1_to_str(num)  # result = result + int1_to_str(num)
            a //= 10  # a = a // 10
        
        result = result[::-1]  # result = '-' * (some_int < 0) + result[::-1]
        
        if number < 0:
            result = '-' + result

    else:
        result = int1_to_str(number)  # result = '0'
    
    return result
````

````python
const = 48  # const = ord('0')

def int1_to_str(number: int) -> str:
    return chr(number + const)

def int_to_str_r(number: int) -> str:
    if number < 10:
        return int1_to_str(number)
    else:
        return int_to_str_r(number // 10) + int1_to_str(number % 10)
````

#### Задача на дом

Есть путь к папке. Надо вывести дерево содержимого этой папки.

````python
# решение задачи к этому занятию

const = 48

def str1_to_int(str_num: str) -> int:
    return ord(str_num) - const

def str_to_int(str_num: str) -> int:
    result = 0
    minus = 0
    
    if str_num[0] == "-":  # .startswith('-')
        str_num = str_num[1:]
        minus = 1

    str_len = len(str_num)

    for i in range(len(str_num)):
        result += str1_to_int(str_num[i]) * (10 ** (str_len - 1 - i))   

    result = (-1) ** minus * result
    
    return result
````
